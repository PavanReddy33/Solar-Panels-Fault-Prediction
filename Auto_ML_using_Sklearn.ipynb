```python
import autosklearn.classification
import numpy as np
import pandas as pd

# Load the data
#data = pd.read_csv("data.csv")

# Split the data into training and test sets
train_data = data.iloc[:int(data.shape[0] * 0.8), :]
test_data = data.iloc[int(data.shape[0] * 0.8):, :]

# Detect and handle outliers in the data
# ...

# Split the data into features and target variables
train_features = train_data.drop("label", axis=1)
train_target = train_data["label"]
test_features = test_data.drop("label", axis=1)
test_target = test_data["label"]

# Convert the data to NumPy arrays
train_features = train_features.values
train_target = train_target.values
test_features = test_features.values
test_target = test_target.values

# Train an auto-sklearn classifier
automl = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=120,
                                                           per_run_time_limit=30,
                                                           seed=1)
automl.fit(train_features, train_target)

# Make predictions with the auto-sklearn classifier
predictions = automl.predict(test_features)

# Evaluate the accuracy of the auto-sklearn classifier
accuracy = np.mean(predictions == test_target)
print("Accuracy:", accuracy)
```
