```python
# Importing all the packages
import pandas as pd
import mysql.connector as sql
```

```python
# Importing the Limited data
df_t = sql.connect(host= "localhost", user="root", password="12345", database= "solarDB",auth_plugin='mysql_native_password')
sql_query = "SELECT * FROM solardb.datasets"
df = pd.read_sql(sql_query,df_t)
```

```python
df
```

```python
df['label'].replace({"F0L": 'NF',"F1L": 'F',"F2L":'F',"F3L": 'F',"F4L": 'F',"F5L": 'F',"F6L": 'F',"F7L": 'F'},inplace=(True))
df['label'].value_counts()
```

```python
# Dropping Unwanted Columns
df = df.drop(df.columns[[0, 1]], axis=1)
df.head()
```

```python
# Importing the Maximum data
df_t1 = sql.connect(host= "localhost", user="root", password="12345", database= "solarDB",auth_plugin='mysql_native_password')
sql_query1 = "SELECT * FROM solardb.datasets_max1"
df1 = pd.read_sql(sql_query1,df_t1)
```

```python
df1['label'].replace({"F0M": 'NF',"F1M": 'F',"F2M":'F',"F3M": 'F',"F4M": 'F',"F5M": 'F',"F6M": 'F',"F7M": 'F'},inplace=(True))
df1['label'].value_counts()
```
